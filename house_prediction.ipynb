```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("housing.csv")
```

```python
data
```

```python
data.info()
```

```python
data.dropna(inplace = True)
# drop na values and saves back the new data
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split
# one needas to define the x and y for the train_test_split to split it to x_train, x_test etc
x = data.drop(['median_house_value'], axis = 1) # the variable is the target value hence we are dropping it and usin gthe rest as independent variables. we ise axis for column
y = data['median_house_value']
```

```python
#data['median_house_value']
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
train_data = x_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist(figsize = (15,8))
```

```python
#plt.figure(figsize=(15,8))
#sns.heatmap(train_data.corr(), annot=True,cmap="YlGnBu")
train_data.corr()
```

```python
train_data['total_rooms'] = np.log(train_data['total_rooms'] +1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms'] +1)
train_data['population'] = np.log(train_data['population'] +1)
train_data['households'] = np.log(train_data['households'] +1)
```

```python
train_data.hist(figsize = (15,8))
```

```python
train_data.ocean_proximity.value_counts()
```

```python
# one hot encoding into binary 0,1
train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity).astype(int)).drop(['ocean_proximity'], axis=1)
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True,cmap="YlGnBu")
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x="latitude", y="longitude", data = train_data, hue="median_house_value", palette="coolwarm")
```

```python
# feature engineering; generating new variables from the existing ones. Can drop features that have an almost zero correlation
train_data['bedroom ratio']=train_data['total_bedrooms']/train_data['total_rooms']
train_data['household_rooms']=train_data['total_rooms']/train_data['households']
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True,cmap="YlGnBu")
```

```python
from sklearn.linear_model import LinearRegression
x_train, y_train = train_data.drop(['median_house_value'], axis = 1), train_data['median_house_value']
reg=LinearRegression()
reg.fit(x_train, y_train)
```

```python
test_data = x_test.join(y_test)

test_data['total_rooms'] = np.log(test_data['total_rooms'] +1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] +1)
test_data['population'] = np.log(test_data['population'] +1)
test_data['households'] = np.log(test_data['households'] +1)

test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity).astype(int)).drop(['ocean_proximity'], axis=1)

test_data['bedroom ratio']=test_data['total_bedrooms']/test_data['total_rooms']
test_data['household_rooms']=test_data['total_rooms']/test_data['households']

x_test, y_test = test_data.drop(['median_house_value'], axis = 1), test_data['median_house_value']
```

```python
from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor()
forest.fit(x_train, y_train)
```

```python
forest.score(x_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
forest = RandomForestRegressor()
param_grid = {
    "n_estimators": [3,10,30],
    "max_features":[2,4,6,8]
}
grid_search = GridSearchCV(forest, param_grid, cv=5,
                           scoring = "neg_mean_squared_error",
                           return_train_score=True)
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_estimator_
```

```python
best_forest = grid_search.best_estimator_
```

```python
best_forest.score(x_test, y_test)
```
